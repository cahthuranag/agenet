```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_age(departure_time, arrival_time, current_time, transmission_error=None):
    if transmission_error is None:
        return current_time - arrival_time
    else:
        return current_time - departure_time

departure_time = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
arrival_time = np.array([2, 7, 12, 17, 22, 27, 32, 37, 42, 47, 52, 57, 62])
current_time = np.array([70 for i in range(len(departure_time))])

age = [calculate_age(departure_time[i], arrival_time[i], current_time[i]) for i in range(len(departure_time))]

plt.plot(arrival_time, age)
plt.xlabel('Time at destination')
plt.ylabel('Age of Information')
plt.title('Age of Information over time at destination')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define time intervals (in seconds)
time = np.linspace(0, 100, 1000)

# Define the AoI equation with transmission errors
def aoi(t):
    return 10 * np.sin(t/10) + 5 * np.sin(t/5) + np.random.normal(0, 1, len(t))

# Calculate the AoI values at each time interval
aoi_values = aoi(time)

# Smooth the AoI values using a rolling average
window_size = 50
smoothed_aoi = np.convolve(aoi_values, np.ones(window_size)/window_size, mode='same')

# Plot the results
plt.plot(time, smoothed_aoi, label='Smoothed AoI')
plt.xlabel('Time (s)')
plt.ylabel('Age of Information (s)')
plt.legend()
plt.show()
```
